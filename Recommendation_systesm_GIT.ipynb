```python
ratings_names = ratings.merge(books , on="ISBN")
ratings_names.head(2)
```

# Getting users with atleast 200 rating

```python
x = ratings_names.groupby('User-ID').count()['Book-Rating'] > 200
users_rating = x[x].index
users_rating
```

# Filtering the rated users book title 

```python
filtered_rating = ratings_names[ratings_names['User-ID'].isin(users_rating)]
filtered_rating.head(5)
```

# **Filtering based on books**

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 70
famous_books =  y[y].index
famous_books
```

```python
final_ratings =  filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

# Filling null values with zero

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
pt
```

# **Calculating cosine similarities**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
scores = cosine_similarity(pt)
scores
```

```python
scores.shape
```

```python
scores = pd.DataFrame(scores)
scores
```

# Creating a function that can take input a book name and returns similar 6 books as recommendation

```python
def recommend(book_name):
    # fetching the index of so and so book 
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(scores[index])), key=lambda x : x[1], reverse=True)[1:7]
    
    for i in similar_items:
        print(pt.index[i[0]]) 
```

```python
print("5 Books Recommended based on this movie:\n")
recommend('4 Blondes')
```

```python
print("5 Books Recommended based on this movie:\n")
recommend('A Bend in the Road')
```
