```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/IMDB-Movie-Data.csv")
```

## **Problem_Statement- EDA on IMDB dataset**

```python
df.head(5)
```

```python
df = df.rename(columns={"Runtime (Minutes)":"Runtime", "Revenue (Millions)":"Revenue"})
df.head()
```

```python
type(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.boxplot(column='Revenue')
plt.show()
```

```python
df.boxplot(column='Metascore')
```

```python
mean_Revenue = df['Revenue'].mean()
mean_Revenue
```

```python
df['Revenue'].fillna(mean_Revenue, inplace = True)
df.head(2)
```

```python
mean_Metascore = df['Metascore'].mean()
mean_Metascore
```

```python
df['Metascore'].fillna(mean_Metascore,inplace = True)
df.head(2)
```

```python
df.isnull().any()
```
