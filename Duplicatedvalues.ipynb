```python
import pandas as pd
```

```python
dup=pd.DataFrame({'col1':['a','a','b','b','b'],'col2':[2,2,2,3,3]})
dup
```

```python
dup1=dup.drop_duplicates()
dup1
```

```python
dup2=dup1.reset_index()
dup2
```

```python
dup3=dup2.drop('index',axis=1)
dup3
```
